```python
#Class Task 2
num=int(input("Enter a number:"))
sum=0
for num in range(1,num+1):
  if num%2==0:
    sum -= num**2
  else:
    sum += num**2
print(sum)
```

```python
#Class Task 4

number = int(input("Enter a number: "))
counter = 0
for i in range(1, number + 1):
   if number % i == 0:
        print(i)
        counter += 1
print("Total"+" " +str(counter)+" "+ "divisors.")
```
