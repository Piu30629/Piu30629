```python
#Home Task 2
str1 = input("Please enter a string: ")
str2 = input("Please enter a string: ")

empty_str = ""
if len(str1) > len(str2):
  smaller = str2
  larger = str1
else:
  smaller = str1
  larger = str2

for p in range(len(smaller)):
  if p < len(smaller):
    empty_str = empty_str + str1[p]
    empty_str = empty_str + str2[p]
  else:
    empty_str = empty_str + larger[p:]

empty_str = empty_str + larger[p+1:]
print(empty_str)
```
