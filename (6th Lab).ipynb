```python
#Task 4
def have_common_member(list1, list2):

    common_flag = False


    for item1 in list1:

        if item1 in list2:
            common_flag = True
            break

    return common_flag
List_one = [1, 4, 3, 2, 6]
List_two = [5, 6, 9, 8, 7]

result = have_common_member(List_one, List_two)
print(result)
```

```python
#Task 7
def nested_to_linear(nested_list):
    if not nested_list:
        return []

    first, *rest = nested_list

    if type(first) == list:
        return nested_to_linear(first) + nested_to_linear(rest)
    else:
        return [first] + nested_to_linear(rest)


nested_input = ["start", 10, [4, 2, [11, [9, "mid", 3, [1, 0], 6]], 8], "Done"]
linear_output = nested_to_linear(nested_input)

print( linear_output)
```

```python
#Task 8
outer_list = []
for i in range(5):
    inp = input()[1:-1]
    name = ""
    num = ""
    flag = True
    for j in inp:
        if j == ',':
            flag = False
        elif flag:
            name += j
        else:
            num += j
    num = float(num)
    outer_list.append([name, num])
    print(outer_list)
```
