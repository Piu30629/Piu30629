```python
#Home Task 2
num = int(input("Enter an number: "))


count = 0
temp = num

while temp > 0:
    temp //= 10
    count+= 1

power = 10 ** (count - 1)

while power > 0:
    digit = num // power
    print(digit, end="," if power >= 10 else "\n" )
    num %= power
    power //= 10


```

```python
#Home Task 3
num = int(input("Enter a number:"))

prime = True
if num <= 1:
    prime = False
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break


sum = 0
for i in range(1, num):
    if num % i == 0:
        sum+=i
if prime:
    print(str(num)+" " +"is a prime number")
elif sum == num:
    print(str(num)+" "+" is a perfect number")
else:
    print(str(num)+" "+"is not a prime or perfect number")
```
