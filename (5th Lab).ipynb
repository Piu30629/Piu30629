```python
#CW-2.2
import math

def tri_area(a, b, c):
  area = 0
  if a+b<=c or b+c<=a or c+a<=b:
    print( "Can’t form triangle")
  else:
    s = (a+b+c)/2
    area = (s*(s-a)*(s-b)*(s-c))**(1/2)

    print(round(area, 3))

tri_area(3,2,1)
```

```python
#CW-3.3
def fitting(d1,d2,dimension):
  if dimension==3:
    volume1=4/3*math.pi*(d1/2)**3
    volume2=4/3*math.pi*(d2/2)**3
    if d1==d2:
      print("Impossible to fit.")
    elif volume1>volume2:
      space1=volume1-volume2
      print("Sphere-2 can fit inside Sphere-1 and",space1,"cubic units would be left")
    else:
      space2=volume2-volume1
      print("Sphere-1 can fit inside Sphere-2 and",space2,"cubic units would be left")
  else:
    area1=math.pi*(d1/2)**2
    area2=math.pi*(d2/2)**2
    if d1==d2:
      print("Impossible to fit.")
    elif area1>area2:
      space3=area1-area2
      print(f"Circle-2 can fit inside Circle-1 and",space3,"square units would be left.")
    else:
      space4=area2-area1
      print("Circle-1 can fit inside Circle-2 and",space4,"square units would be left.")
fitting(8,10,2)
```

```python
#HW-1.1
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
prime_checker=is_prime(15)
print(prime_checker)
```

```python
#HW-1.2
def is_perfect(number):

    if number < 1:
        return False


    divisor_sum = 0
    for i in range(1, number):
        if number % i == 0:
            divisor_sum += i


    return divisor_sum == number

perfect_check = is_perfect(33)
print(perfect_check)
```

```python
#HW-1.3
def special_sum(n):
    total = 0
    for i in range(2, n + 1):
        if is_prime(i) or is_perfect(i):
            total+= i
    return total
result = special_sum(8)
print(result)

```

```python
#HW-5.0
def sequence_iterative(N):
    result = 0
    n = 1

    for i in range(1, N + 1):
        result += n * i
        n *= -1

    return result


result = sequence_iterative(10)
print(result)

```

```python
#HW-5.1
def sequence_recursive(N, n=1):

    if N == 1:
        return 1 * n
    else:

        return N * n + sequence_recursive(N - 1, -n)



result = sequence_recursive(5)
print(result)

```
